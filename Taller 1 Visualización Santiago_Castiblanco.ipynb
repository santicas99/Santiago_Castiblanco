La anterior visualizacion corresponde a la base histórica de rendimientos de cultivos en Colombia de los años 2007 a 2018, relacionada con la producción agrícola Nacional, adicionalmente, brinda a los diversos actores con especial énfasis en los productores, información agrícola regional y nacional que fortalece procesos productivos y de comercialización.

> https://www.datos.gov.co/Agricultura-y-Desarrollo-Rural/RENDIMIENTO-DE-CULTIVOS-EN-COLOMBIA-POR-A-O/fgh5-rjkd

<left>
<h4><span style="font-family: Comic sans; font-size:2em;color:blue"> ¿Qué?</span></h4>
</left>

**Tipo de Dataset:** 
> *Tabla*, al contener items y atributos donde incluso pueden ser interpretados como datos temporales dado que tiene en cuanta una medida de tiempo como los años los cuales harian parte de los items. Ademas la visualizacion en su pagina web viene acompañado de una tabla la cual contiene varios atributos como departamento, municipio, grupo de cultivo, cultivo, entre otros que dados unos items permiten la creacion de los valores de cada año

**Tipo de Datos/Atributos:** 
> *Ordenados* -> **Ordianales** ya que como se observa inicia con el año donde mayor rendimiento hubo hasta le de menor, ordenando de esta forma los años, sin embargo tambien puede ser *Cuantitativos* con una dirección *ciclica* ya que permite identificar que año a año se realizo la misma medición del rendimiento y al ser datos a los que se les puede sacar ejemplo una media, una desviacion, una varianza, entre otras opciones por esta razón tambien son cuantitativos.

**Disponibilidad Dataset:** 
> *Estático*
